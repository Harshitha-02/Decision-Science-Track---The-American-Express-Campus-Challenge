```python
import pandas as pd
import numpy as np

def drop_encode(df):
    # Select the first 23 columns
    df = df.iloc[:, :23]
    
    # Fill NaN values appropriately (use 'missing' for categorical and 0 for numeric columns)
    df = df.fillna({'match id': 'missing', 'team1': 'missing', 'team1_id': 'missing', 'team1_roster_ids': 'missing', 
                    'team2': 'missing', 'team2_id': 'missing', 'team2_roster_ids': 'missing', 'winner': 'missing', 
                    'venue': 'missing', 'city': 'missing', 'match_dt': 'missing', 'series_name': 'missing', 
                    'season': 'missing', 'winner_id': 0, 'toss winner': 'missing', 'toss decision': 'missing',
                    'lighting': 'missing'})
    
    # Ensure all columns in 'drops' exist in df to prevent errors
    drops = ['match id', 'team1', 'team1_id', 'team1_roster_ids', 'team2', 'team2_id', 'team2_roster_ids', 
             'winner_id', 'venue', 'city', 'match_dt', 'series_name', 'season']
    drops = [col for col in drops if col in df.columns]
    
    # Transform 'winner' and 'toss winner' columns based on 'team1'
    df['winner'] = np.where(df['winner'] == df['team1'], 0, 1)
    df['toss winner'] = np.where(df['toss winner'] == df['team1'], 0, 1)
    
    # Transform 'toss decision' to binary
    df['toss decision'] = np.where(df['toss decision'] == 'field', 0, 1)
    
    # One-hot encode the 'lighting' column
    df = pd.get_dummies(df, columns=['lighting'])
    
    # Drop the specified columns
    df = df.drop(drops, axis=1)
    
    # Convert the DataFrame to integers
    df = df.astype(int, errors='ignore')
    
    return df



```

```python
# ground feature enginnering
# Create a feature indicating which team is batting first
match['batting_first'] = np.where((match['toss winner'] == match['team1']) & (match['toss decision'] == 1), 0, 
                                np.where((match['toss winner'] == match['team2']) & (match['toss decision'] == 0), 0, 1))

print("Modified DataFrame with 'batting_first' feature:")
print(match)
match['total_matches_at_ground'] = match['Ground_Batting_win'] + match['Ground_Chasing_win']

# Calculate the win percentage for teams batting first and teams chasing at each ground
match['win_percentage_batting_first'] = (match['Ground_Batting_win'] / match['total_matches_at_ground']) * 100
match['win_percentage_chasing'] = (match['Ground_Chasing_win'] / match['total_matches_at_ground']) * 100
# Set weight to 1 if win percentage batting first is more than 50%, else set weight to 0
match['weight'] = (match['win_percentage_batting_first'] > 50).astype(int)

# Calculate the weighted interaction feature
match['weighted_interaction'] = match['win_percentage_batting_first'] * (1 - match['batting_first']) + \
                                 (100 - match['win_percentage_batting_first']) * match['batting_first']

# print("Modified DataFrame with weighted interaction feature:")
# print(match)
```
